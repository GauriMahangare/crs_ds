```python
imdb_titles['Tokenized_primaryTitle'] = imdb_titles['primaryTitle'].str.lower().apply(nltk.word_tokenize)
```

```python
# Remove punctuation
for index, row in imdb_titles.iterrows():
    text_index = index
    text_words = imdb_titles.at[text_index, 'Tokenized_primaryTitle']
    filtered_text=[]
    for word in text_words:
        if word.isalnum():
            filtered_text.append(word)
    imdb_titles.at[text_index, 'Tokenized_primaryTitle'] = filtered_text
```

```python
# Check if all target words are present in the 'Text' column of df1
for index, row in title.iterrows():
    target_words = row[target_words_column]
    print(target_words)
    # text_index = row['ID'] - 1  # Assuming 'ID' is a unique identifier starting from 1
    text_index = index
    text_words = imdb_titles.at[text_index, 'Tokenized_primaryTitle']
    for i in range(imdb_titles.shape[0]):
        text_words = imdb_titles.at[i, 'Tokenized_primaryTitle']
        if all(word in text_words for word in target_words):
            print(f"For ID {text_index}: All target words {target_words} are present in the text{imdb_titles.at[i, 'primaryTitle']}.")
            print(imdb_titles.at[i,'tconst'])
            title.at[index,'imdb_id'] = imdb_titles.at[i,'tconst']
            break
```

```python
for i in tqdm(range(title.shape[0]),desc="Processing movie titles"):
    title_str = title.loc[i]["title"]
    year_str = title.loc[i]["year"]
    searched_title = imdb_titles[(imdb_titles["primaryTitle"].str.contains(title_str,case=False)) & (imdb_titles["titleType"]=="movie") & (imdb_titles["startYear"].str.contains(year_str,case=False))]
    if searched_title.empty:
        print("Title not found -" + title_str + "-" + year_str)
        continue
    else:
        title.loc[i]["imdb_id"]=searched_title.iloc[0,0]
```
